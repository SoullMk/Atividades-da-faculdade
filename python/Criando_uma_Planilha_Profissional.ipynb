```python
# Instalar o openpyxl (rode uma vez)
!pip install openpyxl -q

import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment
from openpyxl.chart import BarChart, Reference

print('✅ Tudo pronto para criar planilhas!')
```

```python
# Dados de vendas de produtos de tecnologia
vendas = pd.DataFrame({
    'Produto':    ['Mouse',  'Teclado', 'Monitor', 'Webcam', 'Headset', 'MousePad', 'Cabo HDMI'],
    'Quantidade': [150,       80,        45,        120,      95,  200,    60],
    'Preco':      [50,        150,       800,       200,      300, 30, 40],
    'Total':      [7500,      12000,     36000,     24000,    28500, 6000, 2400],
})

print('Dados criados:')
print(vendas)
print(f'\nTotal de vendas: R$ {vendas["Total"].sum():,.2f}')
```

```python
# Criar uma planilha nova
wb = Workbook()
ws = wb.active
ws.title = 'Relatório de Vendas'

# Escrever o cabeçalho (linha 1)
cabecalho = ['Produto', 'Quantidade', 'Preço', 'Total']
ws.append(cabecalho)

# Escrever os dados (linhas 2 em diante)
for _, linha in vendas.iterrows():
    ws.append([linha['Produto'], linha['Quantidade'],
               linha['Preco'],   linha['Total']])

print(f'✅ Dados escritos! A planilha tem {ws.max_row} linhas.')
```

```python
# Cores que vamos usar (código hexadecimal)
COR_FUNDO   = '008000'   # azul escuro
COR_TEXTO   = 'FFFFFF'   # branco

# Percorrer as células da primeira linha e aplicar estilo
for celula in ws[1]:
    celula.font      = Font(bold=True, color=COR_TEXTO, size=12)
    celula.fill      = PatternFill(start_color=COR_FUNDO, fill_type='solid')
    celula.alignment = Alignment(horizontal='center')

# Ajustar a largura das colunas para o conteúdo caber
ws.column_dimensions['A'].width = 15
ws.column_dimensions['B'].width = 14
ws.column_dimensions['C'].width = 12
ws.column_dimensions['D'].width = 14

print('✅ Cabeçalho formatado com sucesso!')
```

```python
# Descobrir em qual linha ficam os totais
ultima_linha = ws.max_row + 1

# Escrever "TOTAL" na coluna A
ws[f'A{ultima_linha}'] = 'TOTAL'
ws[f'A{ultima_linha}'].font = Font(bold=True)

# Fórmula de soma para a coluna Total (coluna D)
ws[f'D{ultima_linha}'] = f'=SUM(D2:D{ultima_linha - 1})'
ws[f'D{ultima_linha}'].font = Font(bold=True)

print(f'✅ Linha de total adicionada na linha {ultima_linha}!')
```

```python
# Criar o gráfico
grafico = BarChart()
grafico.title          = 'Custo de cada produto'
grafico.x_axis.title   = 'Itens comprados'
grafico.y_axis.title   = 'Valor (R$)'
grafico.width          = 18   # largura em cm
grafico.height         = 12   # altura em cm

# Definir os dados: coluna D (Total), linhas 1 a 6
dados_grafico = Reference(ws, min_col=4, min_row=1, max_row=6)
categorias_grafico = Reference(ws, min_col=1, min_row=2, max_row=6)

grafico.add_data(dados_grafico, titles_from_data=True)
grafico.set_categories(categorias_grafico)

# Inserir o gráfico a partir da célula F2
ws.add_chart(grafico, 'F2')

print('✅ Gráfico inserido!')
```

```python
# Salvar o arquivo
nome_arquivo = 'relatorio_vendas_final.xlsx'
wb.save(nome_arquivo)
print(f'✅ Arquivo salvo como "{nome_arquivo}"!')

# Baixar o arquivo no Colab (vai aparecer um download no navegador)
from google.colab import files
files.download(nome_arquivo)
```
