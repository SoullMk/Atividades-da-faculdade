```python
import os
import shutil
from pathlib import Path

print('✅ Bibliotecas prontas!')
```

```python
# Criar a pasta principal
pasta_bagunca = Path('pasta_bagunca')
pasta_bagunca.mkdir(exist_ok=True)   # exist_ok=True = não dá erro se já existir

# Lista de arquivos que vamos criar (nome → extensão)
arquivos_para_criar = [
    'vendas_janeiro.csv',
    'vendas_fevereiro.csv',
    'relatorio_anual.txt',
    'notas_reuniao.txt',
    'analise_clientes.py',
    'modelo_predicao.py',
    'dashboard_config.json',
    'dados_usuarios.json',
    'documentacao.md',
    'README.md',
    'grafico_vendas.csv',
    'script_limpeza.py',
]

# Criar cada arquivo vazio dentro da pasta
for nome in arquivos_para_criar:
    caminho = pasta_bagunca / nome      # "/" com Path = juntar pastas!
    caminho.touch()                     # cria o arquivo vazio
    print(f'📄 Criado: {nome}')

print(f'\n✅ {len(arquivos_para_criar)} arquivos criados em "{pasta_bagunca}"!')
```

```python
# Definir quais extensões vão para qual pasta
categorias = {
    'dados':      ['.csv'],
    'textos':     ['.txt', '.md'],
    'codigos':    ['.py'],
    'configs':    ['.json'],
}

# Criar as subpastas dentro de pasta_bagunca
for nome_pasta in categorias:
    subpasta = pasta_bagunca / nome_pasta
    subpasta.mkdir(exist_ok=True)
    print(f'📁 Pasta criada: {nome_pasta}/')

print('\n✅ Estrutura de pastas criada!')
```

```python
movidos = 0   # contador de arquivos movidos

# Percorrer todos os arquivos da pasta (só arquivos, não subpastas)
for arquivo in pasta_bagunca.iterdir():
    if arquivo.is_file():                    # garantir que é arquivo, não pasta
        extensao = arquivo.suffix            # pega a extensão: .csv, .txt, etc.

        # Descobrir para qual pasta esse arquivo deve ir
        destino = None
        for nome_pasta, extensoes in categorias.items():
            if extensao in extensoes:
                destino = pasta_bagunca / nome_pasta
                break

        # Mover o arquivo se encontramos o destino
        if destino:
            shutil.move(str(arquivo), str(destino / arquivo.name))
            print(f'➡️  {arquivo.name}  →  {destino.name}/')
            movidos += 1

print(f'\n✅ {movidos} arquivos organizados!')
```

```python
# Prefixo que vamos adicionar no nome de todos os arquivos
prefixo = '2026_'

# Percorrer cada subpasta e renomear os arquivos dentro
for nome_pasta in categorias:
    subpasta = pasta_bagunca / nome_pasta

    for arquivo in subpasta.iterdir():
        if arquivo.is_file():
            novo_nome = prefixo + arquivo.name          # '2026_' + 'vendas.csv'
            novo_caminho = arquivo.parent / novo_nome   # mesma pasta, novo nome
            arquivo.rename(novo_caminho)
            print(f'✏️  {arquivo.name}  →  {novo_nome}')

print('\n✅ Todos os arquivos renomeados!')
```

```python
print('=' * 45)
print('       RELATÓRIO DE ORGANIZAÇÃO DE ARQUIVOS')
print('=' * 45)

total_geral = 0

for nome_pasta in categorias:
    subpasta = pasta_bagunca / nome_pasta
    arquivos = list(subpasta.iterdir())
    quantidade = len(arquivos)
    total_geral += quantidade
    print(f'📁 {nome_pasta:<10} → {quantidade} arquivo(s)')

print('-' * 45)
print(f'📊 TOTAL GERAL: {total_geral} arquivos organizados')
print('=' * 45)
```
